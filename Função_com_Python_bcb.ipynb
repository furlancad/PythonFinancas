```python
from bcb import sgs
from bcb import Expectativas
```

```python
em = Expectativas()
```

```python
em.describe()
```

```python
ep = em.get_endpoint('ExpectativasMercadoAnuais')
```

```python
ep.query().filter(ep.Indicador == 'Câmbio').collect()
```

```python
def get_previsoes(indicador, ano):
  return (ep.query()
    .filter(ep.Indicador == f'{indicador}')
    .filter(ep.Data >= f'{ano}-01-01')
    .filter(ep.DataReferencia == ano)
    .select(ep.Data, ep.Mediana)
    .collect()
  )
```

```python
get_previsoes('Câmbio', 2022)
```

```python
get_previsoes('Câmbio', 2022).max()
```

```python
get_previsoes('IPCA', 2022)
```
