```python
ibov2 = pd.DataFrame(index=pd.bdate_range(start="1968-01-01", end="1997-12-31"), columns=["ibov"])
ibov2
```

```python
df = pd.DataFrame()
for ano in range(1968, 1998):
    df = pd.read_excel("IBOVDIA.XLS", index_col=0, header=1, sheet_name=str(ano), nrows=31, usecols="A:M")
    df.columns = range(1,13)
    for mes in df.columns:
      for dia in df.loc[:,mes].dropna().index:
        ibov2.loc[f'{ano}-{mes}-{dia}'] = df.loc[dia,mes]
```

```python
ibov2
```

```python
ibov_recente = yf.download("^BVSP", start="1998-01-01")['Adj Close']
ibov_recente
```

```python
ibov2= pd.concat([ibov2['ibov'], ibov_recente])
ibov2
```

```python
ibov2.plot()
```
